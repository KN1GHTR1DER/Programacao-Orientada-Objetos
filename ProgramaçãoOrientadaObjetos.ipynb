```python
class Veiculo:
  def __init__(self,marca,modelo,preco_base):
    self.marca = marca
    self.modelo = modelo
    self.preco_base = preco_base

  def calcular_valor(self):
    raise NotImplementedError("ERRO! implementar nas subclasses")

class Carro(Veiculo):
  def __init__(self,num_portas,ar_condicionado,marca,modelo,preco_base):
    self.num_portas = num_portas
    self.ar_condicionado = ar_condicionado
    super().__init__(marca,modelo,preco_base)

  def calcular_valor(self):
    if not self.ar_condicionado:
      return self.preco_base*8
    else:
      return self.preco_base*8+5000

class Moto(Veiculo):
  def __init__(self,marca,modelo,preco_base,cilindrada):
    self.cilindrada = cilindrada
    super().__init__(marca,modelo,preco_base)

  def calcular_valor(self):
    if self.cilindrada <= 300:
      return self.preco_base
    elif self.cilindrada >300 and self.cilindrada <=600:
      return self.preco_base*1.25
    else:
      return self.preco_base*2

class Caminhao(Veiculo):
  def __init__(self,marca,modelo,preco_base,capacidade_carga,eixos):
    self.capacidade_carga = capacidade_carga
    self.eixos = eixos
    super().__init__(marca,modelo,preco_base)

  def calcular_valor(self):
    return self.preco_base*10 + self.capacidade_carga*5 + self.eixos*100

class Concessionaria:
  def __init__(self):
    self.estoque = []
    self.veiculos = []

  def incluir_veiculo(self):
    print("Incluir Veículo:")
    marca = input("Marca: ")
    modelo = input("Modelo: ")
    preco_base = float(input("Preço Base: "))
    tipo_veiculo = input("Tipo de Veículo (Carro, Moto, Caminhao): ")
    if tipo_veiculo.lower() == "carro":
      num_portas = int(input("Número de Portas: "))
      ar_condicionado = input("Ar Condicionado (S/N): ").upper() == "S"
      veiculo = Carro(num_portas, ar_condicionado, marca, modelo, preco_base)
    elif tipo_veiculo.lower() == "moto":
      cilindrada = int(input("Cilindrada: "))
      veiculo = Moto(marca, modelo, preco_base, cilindrada)
    elif tipo_veiculo.lower() == "caminhao":
      capacidade_carga = float(input("Capacidade de Carga: "))
      eixos = int(input("Número de Eixos: "))
      veiculo = Caminhao(marca, modelo, preco_base, capacidade_carga, eixos)
    else:
      print("Tipo de Veículo inválido!")
      return
    self.estoque.append(veiculo)
    self.veiculos.append(veiculo)
    print("Veículo incluído com sucesso!")

  def ver_estoque(self):
    print("Estoque:")
    for i, veiculo in enumerate(self.estoque):
      print(f"{i+1}. {veiculo.marca} {veiculo.modelo} - Preço: R$ {veiculo.calcular_preco():.2f}")

  def menu(self):
    while True:
      print("Menu:")
      print("1. Incluir Veículo")
      print("2. Ver Estoque")
      print("3. Sair")
      opcao = input("Opção: ")
      if opcao == "1":
        self.incluir_veiculo()
      elif opcao == "2":
        self.ver_estoque()
      elif opcao == "3":
        break
      else:
        print("Opção inválida!")

concessionaria = Concessionaria()
concessionaria.menu()
```

```python
from abc import ABC, abstractmethod

class Pagamento():
    def __init__(self,quantia,data):
        self.quantia = quantia
        self.data = data

    @abstractmethod
    def Efetuar_pagamento(self):
        pass
    def Quantia(self):
        return f"quantia {self.quantia}, data: {self.data}"

class Dinheiro(Pagamento):
    def __init__(self,quantia,data,moeda):
        self.moeda = moeda
        super().__init__(quantia,data)

    def efetuar_pagamento(self):
        return f"pagamento em {self.moeda} realizado com sucesso"

class Cartao(Pagamento):
    def __init__(self,quantia,data,numero_cartao,validade):
        self.numero_cartao = numero_cartao
        self.validade = validade
        super().__init__(quantia,data)

    def efetuar_pagamemto(self):
        return f"pagamento com cartao {self.numero_cartao} realizado"

class Pix(Pagamento):
    def __init__(self,quantia,data,chave_pix,banco):
        self.chave_pix = chave_pix
        self.banco = banco
        super().__init(quantia,data)

    def efetuar_pagamento(self):
        return f"pagamento da chave {self.chave_pix} do banco{self.banco} realizado"
```

```python
from abc import ABC, abstractmethod

class MaterialBiblioteca():
    def __init__(self, titulo, autor, ano_publicacao):
        self.titulo = titulo
        self.autor = autor
        self.ano_publicacao = ano_publicacao

    @abstractmethod
    def get_informacoes(self):
        pass

class Livro(MaterialBiblioteca):
    def __init__(self,titulo,autor,ano_publicacao)
        super().__init__(titulo,autor,ano_publicacao)

    def get_informacoes(self):
        return f"Livro {self.titulo} escrito por {self.autor} publicado em {self.ano_publicacao}"

class Revista(MaterialBiblioteca):
    def __init__(self,titulo,autor,ano_publicacao)
        super().__init__(titulo,autor,ano_publicacao)

    def get_informacoes(self):
        return f"Revista {self.titulo} escrita por {self.autor} publicada em {self.ano_publicacao}"

class E_book(MaterialBiblioteca):
    def __init__(self,titulo,autor,ano_publicacao)
        super().__init__(titulo,autor,ano_publicacao)

    def get_informacoes(self):
        return f"E-Book {self.titulo} escrito por {self.autor} publicado em {self.ano_publicacao}"

class Audiolivro(MaterialBiblioteca):
    def __init__(self,titulo,autor,ano_publicacao)
        super().__init__(titulo,autor,ano_publicacao)

    def get_informacoes(self):
        return f"Audiolivro {self.titulo} escrito por {self.autor} publicado em {self.ano_publicacao}"

class Biblioteca:
    def __init__(self):
        self.materiais = []

    def adicionar_material(self, material):
        if isinstance(material, MaterialBiblioteca):
            self.materiais.append(material)
        else:
            raise ValueError("O material deve ser uma instância de MaterialBiblioteca")

    def remover_material(self, material):
        if material in self.materiais:
            self.materiais.remove(material)
        else:
            raise ValueError("O material não está na biblioteca")

    def exibir_informacoes(self):
        for material in self.materiais:
            print(material.get_informacoes())
```

```python
from abc import ABC, abstractmethod

class Viagem(ABC):
    def __init__(self, destino, duracao, preco_base):
        self.destino = destino
        self.duracao = duracao
        self.preco_base = preco_base

    @abstractmethod
    def calcular_preco_total(self):
        pass

class Cruzeiro(Viagem):
    def __init__(self, destino, duracao, preco_base, taxa_servico):
        super().__init__(destino, duracao, preco_base)
        self.taxa_servico = taxa_servico

    def calcular_preco_total(self):
        return self.preco_base + (self.preco_base * self.taxa_servico / 100)

class ViagemTerrestre(Viagem):
    def __init__(self, destino, duracao, preco_base, custo_transporte):
        super().__init__(destino, duracao, preco_base)
        self.custo_transporte = custo_transporte

    def calcular_preco_total(self):
        return self.preco_base + self.custo_transporte

class ViagemAerea(Viagem):
    def __init__(self, destino, duracao, preco_base, custo_passagem):
        super().__init__(destino, duracao, preco_base)
        self.custo_passagem = custo_passagem

    def calcular_preco_total(self):
        return self.preco_base + self.custo_passagem

class AgenciaViagens:
    def __init__(self):
        self.viagens = []

    def adicionar_viagem(self, viagem):
        self.viagens.append(viagem)

    def calcular_custo_total(self):
        custo_total = 0
        for viagem in self.viagens:
            custo_total += viagem.calcular_preco_total()
        return custo_total

# Exemplo de uso
agencia = AgenciaViagens()

cruzeiro = Cruzeiro("Caribe", 7, 1000, 10)
viagem_terrestre = ViagemTerrestre("Rio de Janeiro", 5, 800, 200)
viagem_aerea = ViagemAerea("São Paulo", 3, 600, 150)

agencia.adicionar_viagem(cruzeiro)
agencia.adicionar_viagem(viagem_terrestre)
agencia.adicionar_viagem(viagem_aerea)

print("Custo total das viagens:", agencia.calcular_custo_total())
```

```python
def criar_arquivo_com_registros_iniciais(nome_arquivo):
    try:
        with open(nome_arquivo, 'w') as file:
            file.write("Nome: João Silva, cargo: Analista\n")
            file.write("Nome: Maria Souza, cargo: Gerente\n")
        print("Arquivo criado e registros iniciais adicionados com sucesso.")
    except IOError as e:
        print(f"Erro ao criar o arquivo: {e}")

def adicionar_registros(nome_arquivo):
    while True:
        nome = input("Digite o nome do funcionário (ou 'sair' para terminar): ")
        if nome.lower() == 'sair':
            break
        cargo = input("Digite o cargo do funcionário: ")

        try:
            with open(nome_arquivo, 'a') as file:
                file.write(f"Nome: {nome}, cargo: {cargo}\n")
            print("Registro adicionado com sucesso.")
        except IOError as e:
            print(f"Erro ao adicionar o registro: {e}")

def exibir_conteudo_arquivo(nome_arquivo):
    try:
        with open(nome_arquivo, 'r') as file:
            conteudo = file.read()
            print("\nConteúdo do arquivo:")
            print(conteudo)
    except FileNotFoundError:
        print("O arquivo não foi encontrado.")
    except IOError as e:
        print(f"Erro ao ler o arquivo: {e}")

def main():
    nome_arquivo = 'funcionarios.txt'

    # Passo 1: Criar o arquivo com registros iniciais
    criar_arquivo_com_registros_iniciais(nome_arquivo)

    # Passo 2: Adicionar novos registros
    adicionar_registros(nome_arquivo)

    # Passo 3: Exibir o conteúdo do arquivo
    exibir_conteudo_arquivo(nome_arquivo)

if __name__ == "__main__":
    main()
```

```python
import pickle

class Cliente:
  def __init__(self, nome, email):
    self.nome = nome
    self.email = email

class PessoaFisica(Cliente):
  def __init__(self, cpf):
    self.cpf = cpf
    super().__init__(nome,email)

class PessoaJuridica(Cliente):
  def __init__(self,cnpj):
    self.cnpj = cnpj
    super().__init__(nome,email)

class GerenciadorClientes:
  def __init__(self):
    self.pessoas_f = []
    self.pessoas_j = []
    self.carregar_dados()
  def carregar_dados(self):
    try:
      with open('pessoas_fisicas.pkl', 'rb') as f:
        self.pessoas_f = pickle.load(f)
    except (FileNotFoundError, EOFError):
      self.pessoas_f = []

    try:
      with open('pessoas_juridicas.pkl', 'rb') as f:
        self.pessoas_j = pickle.load(f)
    except (FileNotFoundError, EOFError):
      self.pessoas_j = []

  def adicionar_cliente_pf(self):
    nome = input("digite o nome: ")
    email = input("digite o email: ")
    cpf = input("digite o cpf: ")
    cliente_f = (nome,email,cpf)
    self.pessoas_f.append(cliente_f)

  def adicionar_cliente_pj(self):
    nome = input("digite o nome: ")
    email = input("digite o email: ")
    cnpj = input("digite o cnpj: ")
    cliente_j = (nome,email,cnpj)
    self.pessoas_j.append(cliente_j)

  def listar_clientes(self):
    print(self.pessoas_f)
    print(self.pessoas_j)

  def BuscarClientepf(self):
    n = input("digite o indice do cliente a ser buscado: ")
    print(self.pessoas_f[n-1])

  def BuscarClientepj(self):
    n = input("digite o indice do cliente a ser buscado: ")
    print(self.pessoas_j[n-1])

  def ExcluirClientepf(self):
    n = input("digite o indice do cliente a ser excluído: ")
    self.pessoas_f.pop(n-1)

  def ExcluirClientepj(self):
    n = input("digite a posição do cliente a ser excluído: ")
    self.pessoas_j.pop(n-1)

  def Salvar(self):
    with open('pessoas_fisicas.pkl', 'wb') as f:
      pickle.dump(self.pessoas_f, f)
    with open('pessoas_juridicas.pkl', 'wb') as f:
      pickle.dump(self.pessoas_j, f)

def exibir_menu():
    print("Menu:")
    print("1. Adicionar Cliente (Pessoa Física)")
    print("2. Adicionar Cliente (Pessoa Jurídica)")
    print("3. Listar Clientes")
    print("4. Buscar Cliente (Pessoa Física por ID)")
    print("5. Buscar Cliente (Pessoa Jurídica por ID)")
    print("6. Excluir Cliente (Pessoa Física por ID)")
    print("7. Excluir Cliente (Pessoa Jurídica por ID)")
    print("8. Sair e Salvar")

def main():
    gerenciador = GerenciadorClientes()
    while True:
        exibir_menu()
        opcao = input("Digite a opção: ")
        if opcao == "1":
            gerenciador.adicionar_cliente_pf()
        elif opcao == "2":
            gerenciador.adicionar_cliente_pj()
        elif opcao == "3":
            gerenciador.listar_clientes()
        elif opcao == "4":
            #indice = int(input("Digite o índice do cliente a ser buscado: ")) - 1
            gerenciador.BuscarClientepf()
        elif opcao == "5":
            #indice = int(input("Digite o índice do cliente a ser buscado: ")) - 1
            gerenciador.BuscarClientepj()
        elif opcao == "6":
            #indice = int(input("Digite o índice do cliente a ser excluído: ")) - 1
            gerenciador.ExcluirClientepf()
        elif opcao == "7":
            #indice = int(input("Digite o índice do cliente a ser excluído: ")) - 1
            gerenciador.ExcluirClientepj()
        elif opcao == "8":
            gerenciador.Salvar()
            break
        else:
            print("Opção inválida")

if __name__ == "__main__":
    main()
```

```python
 import sqlite3

# Create a connection to the database
conn = sqlite3.connect('banco.db')

# Create a cursor object
cursor = conn.cursor()

# Create the contas table
cursor.execute('''
    CREATE TABLE contas (
        id INTEGER PRIMARY KEY,
        numero_conta TEXT,
        nome_titular TEXT,
        saldo REAL,
        limite_credito REAL
    )
''')

# Commit the changes
conn.commit()

# Close the connection
conn.close()

class ContaBancaria:
    def __init__(self, numero_conta, nome_titular, saldo, limite_credito):
        self.__numero_conta = numero_conta
        self.__nome_titular = nome_titular
        self.__saldo = saldo
        self.__limite_credito = limite_credito

    # Getters and Setters
    def get_numero_conta(self):
        return self.__numero_conta

    def set_numero_conta(self, numero_conta):
        self.__numero_conta = numero_conta

    def get_nome_titular(self):
        return self.__nome_titular

    def set_nome_titular(self, nome_titular):
        self.__nome_titular = nome_titular

    def get_saldo(self):
        return self.__saldo

    def set_saldo(self, saldo):
        self.__saldo = saldo

    def get_limite_credito(self):
        return self.__limite_credito

    def set_limite_credito(self, limite_credito):
        self.__limite_credito = limite_credito

    def save_to_db(self):
        conn = sqlite3.connect('banco.db')
        cursor = conn.cursor()
        cursor.execute('''
            INSERT INTO contas (numero_conta, nome_titular, saldo, limite_credito)
            VALUES (?, ?, ?, ?)
        ''', (self.__numero_conta, self.__nome_titular, self.__saldo, self.__limite_credito))
        conn.commit()
        conn.close()

    @classmethod
    def retrieve_from_db(cls, numero_conta):
        conn = sqlite3.connect('banco.db')
        cursor = conn.cursor()
        cursor.execute('''
            SELECT * FROM contas WHERE numero_conta = ?
        ''', (numero_conta,))
        row = cursor.fetchone()
        if row:
            return cls(*row[1:])
        else:
            return None

def main():
    # Create a new account
    conta = ContaBancaria('123456', 'John Doe', 1000.0, 5000.0)
    conta.save_to_db()

    # Retrieve an account from the database
    retrieved_conta = ContaBancaria.retrieve_from_db('123456')
    if retrieved_conta:
        print(f"Account found: {retrieved_conta.get_nome_titular()} - {retrieved_conta.get_saldo():.2f}")
    else:
        print("Account not found")

if __name__ == '__main__':
    main()

```

```python
import sqlite3
import pickle

class Servico(ABC):
    def __init__(self, tipo_servico, descricao):
        self._tipo_servico = tipo_servico
        self._descricao = descricao

    @property
    def tipo_servico(self):
        return self._tipo_servico

    @property
    def descricao(self):
        return self._descricao

    @abstractmethod
    def retorna_orcamento(self, horas, valor_hora):
        pass

class ServicoInterno(Servico):
    def __init__(self, tipo_servico, descricao, setor, horas):
        super().__init__(tipo_servico, descricao)
        self._setor = setor
        self._horas = horas

    @property
    def setor(self):
        return self._setor

    @property
    def horas(self):
        return self._horas

    def retorna_orcamento(self, horas, valor_hora):
        return (horas * valor_hora) * 0.5

class ServicoExterno(Servico):
    def __init__(self, tipo_servico, descricao, cliente, horas):
        super().__init__(tipo_servico, descricao)
        self._cliente = cliente
        self._horas = horas

    @property
    def cliente(self):
        return self._cliente

    @property
    def horas(self):
        return self._horas

    def retorna_orcamento(self, horas, valor_hora):
        return horas * valor_hora

class GerenciadorServicos:
    def __init__(self, db_file):
        self.conn = sqlite3.connect(db_file)
        self.cursor = self.conn.cursor()
        self.create_tables()

    def create_tables(self):
        self.cursor.execute("""
            CREATE TABLE IF NOT EXISTS PrecoServicos (
                id INTEGER PRIMARY KEY,
                tipo_servico INTEGER,
                valor_hora REAL
            )
        """)
        self.cursor.execute("""
            CREATE TABLE IF NOT EXISTS ServicosInternos (
                id INTEGER PRIMARY KEY,
                tipo_servico INTEGER,
                descricao TEXT,
                setor TEXT,
                horas REAL
            )
        """)
        self.cursor.execute("""
            CREATE TABLE IF NOT EXISTS ServicosExternos (
                id INTEGER PRIMARY KEY,
                tipo_servico INTEGER,
                descricao TEXT,
                cliente TEXT,
                horas REAL
            )
        """)
        self.conn.commit()

    def inserir_servico_interno(self, servico):
        self.cursor.execute("""
            INSERT INTO ServicosInternos (tipo_servico, descricao, setor, horas)
            VALUES (?, ?, ?, ?)
        """, (servico.tipo_servico, servico.descricao, servico.setor, servico.horas))
        self.conn.commit()

    def inserir_servico_externo(self, servico):
        self.cursor.execute("""
            INSERT INTO ServicosExternos (tipo_servico, descricao, cliente, horas)
            VALUES (?, ?, ?, ?)
        """, (servico.tipo_servico, servico.descricao, servico.cliente, servico.horas))
        self.conn.commit()

    def listar_servicos_internos(self):
        self.cursor.execute("SELECT * FROM ServicosInternos")
        return [pickle.loads(row[1]) for row in self.cursor.fetchall()]

    def listar_servicos_externos(self):
        self.cursor.execute("SELECT * FROM ServicosExternos")
        return [pickle.loads(row[1]) for row in self.cursor.fetchall()]

    def buscar_custo_hora_servico(self, tipo_servico):
        self.cursor.execute("SELECT valor_hora FROM PrecoServicos WHERE tipo_servico = ?", (tipo_servico,))
        return self.cursor.fetchone()[0]

# Exemplo
db_file = "servicos.db"
gerenciador = GerenciadorServicos(db_file)

PrecoServicos = {
    1: 100,  # Manutenção básica
    2: 200,  # Reparo avançado
    3: 150,  # Diagnóstico
    4: 250,  # Troca de peças
    5: 50   # Limpeza
}

for tipo_servico, valor_hora in PrecoServicos.items():
    gerenciador.cursor.execute("INSERT INTO PrecoServicos (id, tipo_servico, valor_hora) VALUES (NULL, ?, ?)", (tipo_servico, valor_hora))
gerenciador.conn.commit()

servico_interno = ServicoInterno(1, "Manutenção básica", "Setor 1", 2)
gerenciador.inserir_servico_interno(servico_interno)

servico_externo = ServicoExterno(2, "Reparo avançado", "Cliente 1", 3)
gerenciador.inserir_servico_externo(servico_externo)

print(gerenciador.listar_servicos_internos())  # [ServicoInterno object]
print(gerenciador.listar_servicos_externos())  # [ServicoExterno object]

print(gerenciador.buscar_custo_hora_servico(1))  # 100
print(gerenciador.buscar_custo_hora_servico(2))  # 200
```

```python
import sqlite3
import pickle

class Contato:
    def __init__(self, nome, email, telefone):
        self.__nome = nome
        self.__email = email
        self.__telefone = telefone

    def obter_informacoes(self):
        raise NotImplementedError("Must be implemented by subclass")

class ContatoPessoal(Contato):
    def __init__(self, nome, email, telefone, data_aniversario):
        super().__init__(nome, email, telefone)
        self.__data_aniversario = data_aniversario

    def obter_informacoes(self):
        return f"Nome: {self.__nome}, Email: {self.__email}, Telefone: {self.__telefone}, Data de Aniversário: {self.__data_aniversario}"

class ContatoProfissional(Contato):
    def __init__(self, nome, email, telefone, empresa, cargo):
        super().__init__(nome, email, telefone)
        self.__empresa = empresa
        self.__cargo = cargo

    def obter_informacoes(self):
        return f"Nome: {self.__nome}, Email: {self.__email}, Telefone: {self.__telefone}, Empresa: {self.__empresa}, Cargo: {self.__cargo}"

class GerenciadorContatos:
    def __init__(self, db_file):
        self.conn = sqlite3.connect(db_file)
        self.cursor = self.conn.cursor()
        self.create_table()

    def create_table(self):
        self.cursor.execute("""
            CREATE TABLE IF NOT EXISTS contatos (
                id INTEGER PRIMARY KEY,
                nome VARCHAR(255),
                email VARCHAR(255),
                telefone VARCHAR(20),
                aniversario DATE,
                cargo VARCHAR(255),
                empresa VARCHAR(255),
                tipo VARCHAR(20)
            )
        """)
        self.conn.commit()

    def inserir_contato_pessoal(self, contato):
        self.cursor.execute("""
            INSERT INTO contatos (nome, email, telefone, aniversario, tipo)
            VALUES (?, ?, ?, ?, 'Pessoal')
        """, (contato.nome, contato.email, contato.telefone, contato.data_aniversario))
        self.conn.commit()

    def inserir_contato_profissional(self, contato):
        self.cursor.execute("""
            INSERT INTO contatos (nome, email, telefone, empresa, cargo, tipo)
            VALUES (?, ?, ?, ?, ?, 'Profissional')
        """, (contato.nome, contato.email, contato.telefone, contato.empresa, contato.cargo))
        self.conn.commit()

    def atualizar_contato_pessoal(self, contato):
        self.cursor.execute("""
            UPDATE contatos SET nome = ?, email = ?, telefone = ?, aniversario = ?
            WHERE id = ?
        """, (contato.nome, contato.email, contato.telefone, contato.data_aniversario, contato.id))
        self.conn.commit()

    def atualizar_contato_profissional(self, contato):
        self.cursor.execute("""
            UPDATE contatos SET nome = ?, email = ?, telefone = ?, empresa = ?, cargo = ?
            WHERE id = ?
        """, (contato.nome, contato.email, contato.telefone, contato.empresa, contato.cargo, contato.id))
        self.conn.commit()

    def listar_contatos_pessoais(self):
        self.cursor.execute("SELECT * FROM contatos WHERE tipo = 'Pessoal'")
        return self.cursor.fetchall()

    def listar_contatos_profissionais(self):
        self.cursor.execute("SELECT * FROM contatos WHERE tipo = 'Profissional'")
        return self.cursor.fetchall()

    def listar_contatos(self):
        self.cursor.execute("SELECT * FROM contatos")
        return self.cursor.fetchall()

    def buscar_contato(self, nome):
        self.cursor.execute("SELECT * FROM contatos WHERE nome = ?", (nome,))
        return self.cursor.fetchone()
```

```python
from abc import ABC, abstractmethod
from datetime import datetime

class Veiculo(ABC):
  quantidade_veiculos = 0

  def __init__(self,modelo,ano,valor):
    self.modelo = modelo
    self.ano = ano
    self.valor = valor
    Veiculo.quantidade_veiculos +=1

  @staticmethod
  def validar_ano(ano):
    ano_atual = datetime.now().year
    return 1886<=ano<=ano_atual

  @classmethod
  def total_veiculos(cls):
    return cls.quantidade_veiculos

  @abstractmethod
  def calcular_seguro(self):
    pass

class Carro(Veiculo):
  def calcular_seguro(self):
    return self.valor * 0.05

class Moto(Veiculo):
  def calcular_seguro(self):
    return self.valor * 0.03

class Caminhao(Veiculo):
  def calcular_seguro(self):
    return self.valor * 0.10
```

```python
class Pessoa:
    def __init__(self, nome, cpf, data_nascimento):
        self.nome = nome
        self.cpf = cpf
        self.data_nascimento = data_nascimento


class Paciente:
    def __init__(self, pessoa):
        self.pessoa = pessoa


class Medico:
    def __init__(self, pessoa, crm, especialidade):
        self.pessoa = pessoa
        self.crm = crm
        self.especialidade = especialidade


class Consulta:
    def __init__(self, paciente, medico, data_hora, observacoes):
        self.paciente = paciente
        self.medico = medico
        self.data_hora = data_hora
        self.observacoes = observacoes


class Menu:
    def __init__(self):
        self.medicos = {}
        self.pacientes = {}
        self.consultas = []

    def incluir_medico(self):
        nome = input("Nome do médico: ")
        cpf = input("CPF do médico: ")
        data_nascimento = input("Data de nascimento do médico (dd/mm/aaaa): ")
        crm = input("CRM do médico: ")
        especialidade = input("Especialidade do médico: ")

        pessoa = Pessoa(nome, cpf, data_nascimento)
        medico = Medico(pessoa, crm, especialidade)
        self.medicos[crm] = medico
        print("Médico cadastrado com sucesso!")

    def incluir_paciente(self):
        nome = input("Nome do paciente: ")
        cpf = input("CPF do paciente: ")
        data_nascimento = input("Data de nascimento do paciente (dd/mm/aaaa): ")

        pessoa = Pessoa(nome, cpf, data_nascimento)
        paciente = Paciente(pessoa)
        self.pacientes[cpf] = paciente
        print("Paciente cadastrado com sucesso!")

    def agendar_consulta(self):
        cpf_paciente = input("Digite o CPF do paciente: ")
        crm_medico = input("Digite o CRM do médico: ")
        data_hora = input("Data e hora da consulta (dd/mm/aaaa HH:MM): ")
        observacoes = input("Observações: ")

        paciente = self.pacientes.get(cpf_paciente)
        medico = self.medicos.get(crm_medico)

        if paciente and medico:
            consulta = Consulta(paciente, medico, data_hora, observacoes)
            self.consultas.append(consulta)
            print("Consulta agendada com sucesso!")
        else:
            print("Paciente ou médico não encontrado!")

    def ver_agenda(self):
        if not self.consultas:
            print("Não há consultas agendadas.")
            return

        for consulta in self.consultas:
            print(f"Consulta: {consulta.data_hora} - Paciente: {consulta.paciente.pessoa.nome} - Médico: {consulta.medico.pessoa.nome} - Observações: {consulta.observacoes}")

    def exibir_menu(self):
        while True:
            print("\nMenu:")
            print("1. Incluir Médico")
            print("2. Incluir Paciente")
            print("3. Agendar Consulta")
            print("4. Ver Agenda")
            print("5. Sair")

            opcao = input("Escolha uma opção: ")
            if opcao == '1':
                self.incluir_medico()
            elif opcao == '2':
                self.incluir_paciente()
            elif opcao == '3':
                self.agendar_consulta()
            elif opcao == '4':
                self.ver_agenda()
            elif opcao == '5':
                print("Saindo do sistema.")
                break
            else:
                print("Opção inválida! Tente novamente.")


if __name__ == "__main__":
    menu = Menu()
    menu.exibir_menu()
```

```python
class CadastroUsuario:
  def __init__(self, nome, cpf, nascimento, endereco, telefone, senha):
    self.nome = nome
    self.cpf = cpf
    self.nascimento = nascimento
    self.endereco = endereco
    self.telefone = telefone
    self.senha = senha

  def mostrar_info():
    print(f"Nome: {self.nome}")
    print(f"CPF: {self.cpf}")
    print(f"Nascimento: {self.nascimento}")
    print(f"Endereço: {self.endereco}")
    print(f"Telefone: {self.telefone}")

  def gravar_info():
    if self.validar_gravacao:
      with open(f'cadastro_{self.cpf}.txt', 'w') as arquivo:
        arquivo.write(f'Nome: {self.nome}\n')
        arquivo.write(f'CPF: {self.cpf}\n')
        arquivo.write(f'Nascimento: {self.nascimento}\n')
        arquivo.write(f'Endereço: {self.endereco}\n')
        arquivo.write(f'Telefone: {self.telefone}\n')
      print(f"Arquivo cadastro {self.cpf}.txt realizado com sucesso!")
    else:
      print("Erro ao realizar gravação!")

  def validar_nome():
    return len(self.nome.split()) >= 2 and self.nome.strip != ""

  def validar_cpf():
    return len(self.cpf) == 14 and self.cpf.count(".") == 2 and self.cpf.count("-") == 1

  def validar_nascimento():
    dia, mes, ano = self.nascimento.split("/")
    return dia.isdigit() and mes.isdigit() and ano.isdigit() and int(dia) <= 31 and int(mes) <= 12 and int(ano) <= 2022

  def validar_endereco():
    return self.endereco.strip() != ""

  def validar_telefone():
    return len(self.telefone) in [10, 11] and self.telefone.count("(") == 1 and self.telefone.count(")") == 1 and self.telefone.count("-") == 1

  def validar_senha():
    return len(self.senha) >= 8 and any(char.isdigit() for char in self.senha) and any(char.isalpha() for char in self.senha)

  def validar_gravacao():
    if not self.validar_nome():
      print("Nome inválido!")
      return False
    if not self.validar_cpf():
      print("CPF inválido!")
      return False
    if not self.validar_nascimento():
      print("Data de nascimento inválida!")
      return False
    if not self.validar_endereco():
      print("Endereço inválido!")
      return False
    if not self.validar_telefone():
      print("Telefone inválido!")
      return False
    if not self.validar_senha():
      print("Senha inválida!")
      return False

    return True
```
